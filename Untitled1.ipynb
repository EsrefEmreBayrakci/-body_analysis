```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from tensorflow.keras.layers import Dropout
```

```python
dataFrame = pd.read_excel("kitap1.xlsx")
dataFrame
```

```python
dataFrame.describe()
```

```python
dataFrame.corr(numeric_only=True)
```

```python
y = dataFrame["ic_organ_yaglanma"].values
x = dataFrame.drop(["vucut_yağ_orani", "vucut_sivi_orani", "ic_organ_yaglanma", "kemik_agirligi", "kas_kutlesi"], axis=1).values
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.05, random_state=15)
```

```python
scaler = MinMaxScaler()
```

```python
x_train = scaler.fit_transform(x_train)
x_test = scaler.transform(x_test)
```

```python
model = Sequential()
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(16, activation='relu'))

model.add(Dense(1))

model.compile(optimizer="adam", loss="mse")
```

```python
model.fit(x=x_train, y=y_train, validation_data=(x_test, y_test), epochs=200)
```

```python
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
y_pred = model.predict(x_test)
```

```python
# Mean Squared Error (Ortalama Kare Hatası) ile model performansını değerlendir
mse = mean_squared_error(y_test, y_pred)
mne = mean_absolute_error(y_test, y_pred)
```

```python
y_pred
```

```python
mse
```

```python
mne
```

```python
data = np.array([60,166,63,99,31])
tahmin = scaler.transform(data.reshape(-1,5))
newpredict = model.predict(tahmin)
```

```python
newpredict
```
